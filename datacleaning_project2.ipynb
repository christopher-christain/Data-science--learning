```python
# this is used to check if Book Table contain ther values aside from yes and no
# NOTE THE YES AND NO ARE CASE SENSITIVE
invalid_values = data[~data['Book Table'].isin(['Yes', 'No'])]

if invalid_values.empty:
    print("✅ Column is clean — contains only 'Yes' and 'No'")
else:
    print("❌ Column contains other values:")
    print(invalid_values['Book Table'].unique())
```
