```python
import pandas as pd
```

# Calculate the mean value for the normalized column

```python
#finding the mean of the column and making sure all the numbers are treated as digits and not strings 

avg_norm_loss = data["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

# Replace NAN value with the mean values

```python
data["normalized-losses"]=data["normalized-losses"].replace(np.nan, avg_norm_loss)
```

```python
data["normalized-losses"]
```

```python
# CHECKING FOR NULL VALUE IN COLUMN "STROKE"
data['stroke'].isnull().sum()
```

```python
#calculate the mean value for th normalized column of stroke

avg_stroke = data["stroke"].astype("float").mean(axis = 0)
print ("Average of stroke",avg_stroke)
```

```python
# replace the NAN value with the average of stroke 

data['stroke'].replace(np.nan, avg_stroke)
```

```python
# save to the data

data["stroke"] = data["stroke"].replace(np.nan, avg_stroke)
```
