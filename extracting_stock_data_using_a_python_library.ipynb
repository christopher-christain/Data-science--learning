```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# install all needed python library
```

```python
# apple stock data extraction from json

apple = yf.Ticker("AAPL")
```

```python
!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/data/apple.json
```

```python
import json
with open('apple.json') as json_file:
    apple_info = json.load(json_file)
apple_info
```

```python
apple_info['country']
```

```python
# extract apple share data

apple_share_data = apple.history(period = 'max')
apple_share_data.head()
```

```python
apple_share_data.reset_index(inplace = True)
```

```python
apple_share_data
```

```python
apple_share_data.plot(x="Date", y="Open")
```

```python
apple.dividends
```

```python
apple.dividends.plot(x='Date', y="Dividends")
```
